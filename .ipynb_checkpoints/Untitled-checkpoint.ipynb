```python
import pandas as pd
import matplotlib as mpl 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Merged.csv')
```

```python
df.head()
```

```python
x = df.groupby('college')['name'].count().sort_values(ascending = False).index[1:10]
```

```python
x
```

```python
import seaborn as sns
```

```python
sns.countplot(data = df[(df['college'].isin(x)) & (df['YEAR'] > 2010)],
             x = 'YEAR',
             hue = 'college',
)
```

```python
a = pd.crosstab(df[df['YEAR'] > 2000]['YEAR'], df[df['college'].isin(x)].college)
```

```python
b = x.tolist()
```

```python
fig, ax = plt.subplots()

ax.plot(a, label = b)
ax.legend(loc = 'best')
```
